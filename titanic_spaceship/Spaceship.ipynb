# SVM 

```python
import pandas as pd
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.impute import SimpleImputer

# Charger les données d'entraînement et de test
train = pd.read_csv("train_spaceship.csv")
test = pd.read_csv("test_spaceship.csv")

# Convertir les colonnes booléennes en valeurs binaires (0 et 1)
train["CryoSleep"] = train["CryoSleep"].replace({False: 0, True: 1})
train["Transported"] = train["Transported"].replace({False: 0, True: 1})
train["VIP"] = train["VIP"].replace({False: 0, True: 1})
train["HomePlanet"] = train["HomePlanet"].replace({'Europa': 0, 'Earth': 1, 'Mars': 2})
train["Destination"] = train["Destination"].replace({'TRAPPIST-1e': 0, '55 Cancri e': 1, 'PSO J318.5-22': 2})

# Supprimer les colonnes inutiles
train = train.drop(["Name", "Cabin"], axis=1)
test = test.drop(["Name", "Cabin"], axis=1)

# Séparer les caractéristiques (X) de la cible (y)
X_train = train.drop("Transported", axis=1)
y_train = train["Transported"]

# Imputer les valeurs manquantes
imputer = SimpleImputer(strategy="mean")
X_train = imputer.fit_transform(X_train)
test = imputer.transform(test)

# Normaliser les caractéristiques (ce qui est important pour les SVM)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(test)

# Créer et entraîner le modèle SVM
svm_classifier = SVC(kernel='linear')  # Vous pouvez choisir le noyau approprié (linéaire, RBF, etc.)
svm_classifier.fit(X_train, y_train)

# Prédire les étiquettes sur les nouvelles données
test_predictions = svm_classifier.predict(X_test)

# Les prédictions sont stockées dans test_predictions, que vous pouvez utiliser pour vos besoins.
```
